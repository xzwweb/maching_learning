```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data_path = './orl_faces'
class_names = os.listdir(data_path)
Images = []

dict_label2ID = {
    's1': 0,
    's2': 1,
    's3': 2,
    's4': 3,
    's5': 4,
    's6': 5,
    's7': 6,
    's8': 7,
    's9': 8,
    's10':9,
    's11': 10,
    's12': 11,
    's13': 12,
    's14': 13,
    's15': 14,
    's16': 15,
    's17': 16,
    's18': 17,
    's19': 18,
    's20': 19,
    's21': 20,
    's22': 21,
    's23': 22,
    's24': 23,
    's25': 24,
    's26': 25,
    's27': 26,
    's28': 27,
    's29': 28,
    's30': 29,
    's31': 30,
    's32': 31,
    's33': 32,
    's34': 33,
    's35': 34,
    's36': 35,
    's37': 36,
    's38': 37,
    's39': 38,
    's40': 39
}

for class_name in class_names:  # 对于每个类别
    class_dir = os.path.join(data_path, class_name)  # 获取类别对应的文件夹路径
    file_names = os.listdir(class_dir)  # 在该文件夹下，获取所有图片文件名
    label_id = dict_label2ID.get(class_name)
    
    n_sample = len(file_names)
    for i in range(n_sample):
        file_name = file_names[i]
        image_path = os.path.join(class_dir, file_name)  # 获取每个图片的路径
        Images.append([image_path, label_id])
```

```python
for i in range(400):
    if Images[i][1] == 0:
        img = plt.imread(Images[i][0])
        plt.imshow(img)
        plt.show()
```
